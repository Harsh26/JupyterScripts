```python
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

```python
train_labels = []
train_samples = []
```

```python
for i in range(50):
    # 5% Younger people who experienced Side-Effect
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(1)
    
    # 5% Older people who DID NOT experienced Side-Effect
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(0)
```

```python
for i in range(1000):
    # 95% Younger people who DID NOT experienced Side-Effect
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(0)
    
    # 95% Older people who experienced Side-Effects
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(1)
```

```python
for i in train_samples:
    print(i)
```

```python
for i in train_labels:
    print(i)
```

```python
train_labels=np.array(train_labels)
train_samples=np.array(train_samples)
train_labels,train_samples=shuffle(train_labels,train_samples)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_samples=scaler.fit_transform(train_samples.reshape(-1,1))
```

```python
for i in scaled_train_samples:
    print(i)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

```python
#physical_devices=tf.config.experimental.list_physical_devices('GPU')
#print("Number of GPUs Available: ",len(physical_devices))
#tf.config.experimental.set_memory_growth(physical_devices[0],True)
```

```python
model = Sequential([
                        Dense(units=16,input_shape=(1,),activation='relu'),
                        Dense(units=32,activation='relu'),
                        Dense(units=2,activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.0001),loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
# Training the NeuralNetwork
model.fit(x=scaled_train_samples, y=train_labels, validation_split=0.1, batch_size=10, epochs=30, shuffle=True, verbose=2)
```

```python
# Test Data Generation
test_labels=[]
test_samples=[]
```

```python
for i in range(10):
    # 5% young people who got side-effects
    random_younger = randint(13,64)
    test_samples.append(random_younger)
    test_labels.append(1)
    
    # 5% elders who NOT got side-effects
    random_older= randint(65,100)
    test_samples.append(random_older)
    test_labels.append(0)
```

```python
for i in range(200):
    # 95% young people who DID NOT get side-effects
    random_younger = randint(13,64)
    test_samples.append(random_younger)
    test_labels.append(0)
    
    # 95% Elders who got side-effects
    random_older = randint(65,100)
    test_samples.append(random_older)
    test_labels.append(1)
    
#print(len(test_samples),len(test_labels),sep=" ")
    
```

```python
test_labels = np.array(test_labels)
test_samples = np.array(test_samples)
test_labels, test_samples = shuffle(test_labels,test_samples)
```

```python
scaled_test_samples = scaler.fit_transform(test_samples.reshape(-1,1))
```

```python
for i in scaled_train_samples:
    print(i)
    
```

# Prediction 

```python
predictions = model.predict(x=scaled_test_samples,batch_size=10,verbose=2)
```

```python
for i in predictions:
    print (i)
```

```python
rounded_predictions = np.argmax(predictions,axis=-1)
```

```python
for i in rounded_predictions:
    print(i)
```

# Confusion Matrix. Since we are the ones to generate test data

```python
%matplotlib inline
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import itertools
import matplotlib.pyplot as plt
```

```python
cm = confusion_matrix(y_true = test_labels,y_pred = rounded_predictions)
```

```python
def plotConfusionMatrix(cm,classes):
    disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels = classes)
    disp.plot()
    disp.ax_.set_title("Confusion Matrix")
    plt.show()
```

```python
plotConfusionMatrix(cm,["no side effects","side effects"])
```
